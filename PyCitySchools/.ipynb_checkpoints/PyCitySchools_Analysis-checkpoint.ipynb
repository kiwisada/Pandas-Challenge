```python
import pandas as pd

schools_file = "Resources/schools_complete.csv"
students_file = "Resources/students_complete.csv"

schools_file_df = pd.read_csv(schools_file)
students_file_df = pd.read_csv(students_file)

all_school_data = pd.merge(schools_file_df, students_file_df, how="left", on=["school_name", "school_name"])
all_school_data.head()
```

```python
school_count = len(all_school_data["school_name"].unique())
school_count
```

```python
student_count = all_school_data["school_name"].count()
student_count
```

```python
grouped_budget_df = all_school_data["budget"].unique()
total_budget = grouped_budget_df.sum()
total_budget
```

```python
average_math_score = all_school_data["math_score"].mean()
average_math_score
```

```python
average_reading_score = all_school_data["reading_score"].mean()
average_reading_score
```

```python
passing_math_count = all_school_data[(all_school_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
passing_reading_count = all_school_data[(all_school_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
passing_math_reading_count = all_school_data[
    (all_school_data["math_score"] >= 70) & (all_school_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
district_summary = pd.DataFrame({"Total Number of Unique Schools":[school_count], "Total Students": [student_count],
                                "Total Budget":[total_budget], "Average Math Score":[average_math_score],
                                 "Average Reading Score":[average_reading_score],
                                "% Passing Math": [passing_math_percentage],
                                 "% Passing Reading": [passing_reading_percentage],
                                "% Overall Passing":[overall_passing_rate]})

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

district_summary
```

```python
school_types = schools_file_df.set_index(["school_name"])["type"]
school_types
```

```python
per_school_counts = all_school_data["school_name"].value_counts()
per_school_counts
```

```python
per_school_budget = all_school_data.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_counts
```

```python
per_school_math = all_school_data.groupby(["school_name"]).mean()["math_score"]
per_school_reading = all_school_data.groupby(["school_name"]).mean()["reading_score"]
per_school_reading
```

```python
school_passing_math = all_school_data[all_school_data["math_score"] >= 70]
school_passing_math
```

```python
school_passing_reading = all_school_data[all_school_data["reading_score"] >= 70]
school_passing_reading
```

```python
passing_math_and_reading = all_school_data[
    (all_school_data["reading_score"] >= 70) & (all_school_data["math_score"] >= 70)]
passing_math_and_reading
```

```python
per_school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
per_school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
overall_passing_rate
```

```python
per_school_summary = pd.DataFrame({"Total school budget":per_school_budget, "Per student budget":per_school_capita,
                                  "Average math score": per_school_math, "Average reading score": per_school_reading,
                                  "% passing math": per_school_passing_math, "% passing reading": per_school_passing_reading,
                                  "% overall passing": overall_passing_rate})

per_school_summary["Total school budget"] = per_school_summary["Total school budget"].map("${:,.2f}".format)
per_school_summary
```

```python
top_schools = per_school_summary.sort_values("% overall passing", ascending=False)
top_schools.head(5)
```

```python
top_schools = per_school_summary.sort_values("% overall passing", ascending=True)
top_schools.head(5)
```

```python
ninth_graders = all_school_data[(all_school_data["grade"] == "9th")]
tenth_graders = all_school_data[(all_school_data["grade"] == "10th")]
eleventh_graders = all_school_data[(all_school_data["grade"] == "11th")]
twelfth_graders = all_school_data[(all_school_data["grade"] == "12th")]

ninth_graders_scores = ninth_graders.groupby(["school_name"])["math_score","reading_score"].mean()
tenth_graders_scores = tenth_graders.groupby(["school_name"])["math_score","reading_score"].mean()
eleventh_graders_scores = eleventh_graders.groupby(["school_name"])["math_score","reading_score"].mean()
twelfth_graders_scores = twelfth_graders.groupby(["school_name"])["math_score","reading_score"].mean()

ninth_grade_math_scores = ninth_graders_scores["math_score"]
tenth_grader_math_scores = tenth_graders_scores["math_score"]
eleventh_grader_math_scores = eleventh_graders_scores["math_score"]
twelfth_grader_math_scores = twelfth_graders_scores["math_score"]

math_scores_by_grade = pd.DataFrame({"9th":ninth_grade_math_scores,
                                     "10th":tenth_grader_math_scores,
                                    "11th":eleventh_grader_math_scores,
                                    "12th":twelfth_grader_math_scores})
math_scores_by_grade = math_scores_by_grade[["9th", "10th", "11th", "12th"]]
math_scores_by_grade.index.name = None
math_scores_by_grade
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-685"]
```

```python
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels = labels)
per_school_summary
```

```python
spending_math_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average math score"]
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average reading score"]
spending_passing_math = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% passing math"]
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% passing reading"]
overall_passing_spending = per_school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% overall passing"]
```

```python
spending_summary = pd.DataFrame({"Average math score":spending_math_scores, "Average reading score":spending_reading_scores,
                                "% passing math":spending_passing_math, "% passing reading":spending_passing_reading,
                                "% overall passing":overall_passing_spending})
spending_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
per_school_summary["School Size"] = pd.cut(per_school_counts, size_bins, labels = labels)
per_school_summary
```

```python
size_math_scores = per_school_summary.groupby(["School Size"]).mean()["Average math score"]
size_reading_scores = per_school_summary.groupby(["School Size"]).mean()["Average reading score"]
size_passing_math = per_school_summary.groupby(["School Size"]).mean()["% passing math"]
size_passing_reading = per_school_summary.groupby(["School Size"]).mean()["% passing reading"]
size_overall_passing = per_school_summary.groupby(["School Size"]).mean()["% overall passing"]
```

```python
size_summary = pd.DataFrame({"Average math score":size_math_scores, "Average reading score":size_reading_scores,
                                "% passing math":size_passing_math, "% passing reading":size_passing_reading,
                                "% overall passing":size_overall_passing})
size_summary
```
